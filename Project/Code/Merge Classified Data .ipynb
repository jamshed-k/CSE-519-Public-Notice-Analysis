```python
df_com = pd.merge(df_test_1, df_test_2, on='Unnamed: 0')
```

```python
len(df_com)
```

```python
df_com.tail(50)
```

```python
df_com = df_com[['state_x','county_x','newspaper_x','date_x','content_x','category_x','year_x', 'month_x', 'category_y']]
```

```python
df_com.head()
```

```python
def merge_category(row):
    cat1 = row[5]
    cat2 = row[8]
    
    if cat1=='Other':
        return cat2
    else:
        return cat1
```

```python
df_com['category'] = df_com.apply(merge_category, axis=1)
```

```python
df_com.head()
```

```python
df_com = df_com.drop(['category_x','category_y'], axis=1)
```

```python
df_com = df_com.rename(columns={'state_x': 'state', 'county_x': 'county',
                               'newspaper_x': 'newspaper', 'date_x': 'date',
                               'content_x': 'content', 'year_x': 'year',
                               'month_x': 'month'})
```

```python
df_com.head()
```

```python
df_temp = pd.read_csv('../Data/all_combined.csv')
print(len(df_temp))
```

```python
df_temp.head()
```

```python
df_temp = df_temp[['state','county','newspaper','date','content','category_no_duplicate']]
```

```python
df_temp.head()
```

```python
df_temp.content = df_temp.content.str.lower()
df_temp.content = df_temp.content.str.replace('\n',' ')
df_temp.content = df_temp.content.str.replace('\r',' ')
```

```python
df_temp['year'] = df_temp['date'].str[:4]
df_temp['year'] = df_temp['year'].astype(int)

df_temp['month'] = df_temp['date'].str[5:7]
df_temp['month'] = df_temp['month'].astype(int)
```

```python
df_temp.head()
```

```python
df_temp = df_temp.rename(columns={'category_no_duplicate': 'category'})
```

```python
df_temp.category = df_temp.category.str.lower()
df_temp.category = df_temp.category.str.replace('[','')
df_temp.category = df_temp.category.str.replace(']','')
df_temp.category = df_temp.category.str.replace('\'','')
```

```python
df_temp.head()
```

```python
df_temp = df_temp.append(df_com)
```
